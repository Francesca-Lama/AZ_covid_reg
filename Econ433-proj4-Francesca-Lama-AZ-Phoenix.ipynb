### 2b 
The 3 COVID related policies I am choosing to focus on are as follows:
* 3/20/20 - the stay-at-home order, which I call 'b_shutdown' and 'a_shutdown' to signify times before and after COVID
* 8/10/20 - more non-essential businesses reopened, 'reopen'

```python
# creating new variables regarding to Arizona COVID regulations
azMerged['b_shutdown'] = np.where(azMerged['date'] <= '2020-03-20', 1, 0)
azMerged['a_shutdown'] = np.where(azMerged['date'] > '2020-03-20', 1, 0)
azMerged['reopen'] = np.where(azMerged['date'] >= '2020-08-10', 1, 0)
```

I narrowed down the list of variables I will be using in my regression based on the ones I did not think would be useful in predicting y ~ 'log(dailyvisits+1)'.

```python
# new list of variables to predict log(dailyvisits+1)
features = azMerged.loc[:,['weekend','region','bizgroup','dayname','month',
                               'B01001e1','B01002e1','B11001e1',
                               'B11001e2','B11001e3',
                               'EDU_higher','B19013e1','B19025e1',
                               'B19301e1','TAVG','cases','deaths',
                               'b_shutdown','a_shutdown']]
```

```python
# seperating Local Merged into a training sample and test sample
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features,y,
                                                    random_state = 1)
```
